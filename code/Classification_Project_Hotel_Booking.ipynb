````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.metrics import precision_score, recall_score, f1_score,accuracy_score,confusion_matrix,make_scorer
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
import pickle
````

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier,AdaBoostClassifier,StackingClassifier,VotingClassifier
import xgboost as xgb
from mlxtend.classifier import StackingClassifier
````

###  Splitting data into training and testing 

````python
y = df_model.is_canceled
X = df_model.drop(columns='is_canceled')
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= .2, random_state= 0)
len(X_train),len(X_test)
````

### Class Imbalance check

````python
df_model.is_canceled.value_counts()
````

From above numbers, we can see that the negative and positive observations are relatively balanced, further class imblance technique might not be needed.

## 1. Baseline models

##### Scale date

````python
scaled = StandardScaler()
X_train_scaled = scaled.fit_transform(X_train)
X_test_scaled = scaled.transform(X_test)
````

##### Model list

````python
knn = KNeighborsClassifier(n_neighbors=5)
logit = LogisticRegression(max_iter = 1000)
decision_tree = DecisionTreeClassifier()
random_forest = RandomForestClassifier()
extra_tree = ExtraTreesClassifier()
adaboost = AdaBoostClassifier()
gbm = xgb.XGBClassifier()
bernoullinb = BernoulliNB()
gaussianb = GaussianNB()
````

````python
model_names = ['knn','logit','decision_tree','random_forest','extra_tree','adaboost','gbm','bernoullinb','gaussianb']
model_vars = [eval(n) for n in model_names]
model_list = list(zip(model_names, model_vars))
````

````python
f1_train_list,accuracy_list,precision_list,recall_list,f1_test_list,confusion_list=[],[],[],[],[],[]
````

````python
for model in model_names:
    curr_model = eval(model)
    curr_model.fit(X_train, y_train)
    print(f"{model}:  {curr_model}")
    y_train_pred = curr_model.predict(X_train)
    y_test_pred = curr_model.predict(X_test)
    with open(f"{model}.pickle","wb")as pfile:
        pickle.dump(curr_model,pfile)
    f1_train = f1_score(y_train, y_train_pred)
    accuracy = accuracy_score(y_test, y_test_pred)
    precision = precision_score(y_test, y_test_pred)
    recall = recall_score(y_test, y_test_pred)
    f1_test = f1_score(y_test, y_test_pred)
    confusion = confusion_matrix(y_test,y_test_pred)
    
    #roc_auc_score = roc_auc_score(y_test,y_test_pred)
    #fpr_model, tpr_model, thresholds_ = roc_curve(y_test, curr_model.predict_proba(X_test_scaled)[:,1])
    
    f1_train_list.append(f1_train)
    accuracy_list.append(accuracy)
    precision_list.append(precision)
    recall_list.append(recall)
    f1_test_list.append(f1_test)
    confusion_list.append(confusion)
    #roc_auc_score_list.append(roc_auc_score)
    
    print("f1 score of training data is:",f1_train)
    print("accuracy score of test data is:",accuracy)
    print("precision score of test data is:",precision)
    print("recall score of test data is:",recall)
    print("f1 score of test bdata is:",f1_test)
    print("ROC AUC score = ",roc_auc_score)
    print("confusion matrix of test data is:\n",confusion)
    
    print("\n ---------------------------------------------------------------------\n")
````

````python
basic_result = pd.DataFrame(data = [f1_train_list,accuracy_list,precision_list,recall_list,f1_test_list]).T
basic_result.columns = ['f1_train','accuracy','precision','recall','f1_test']
basic_result.index = model_names
basic_result
````

#####  ROC AUC

````python
roc_auc_score_list = []
y_train_pred = logit.predict(X_train_scaled)
y_test_pred = logit.predict(X_test_scaled)
fpr_logit, tpr_logit, thresholds = roc_curve(y_test, logit.predict_proba(X_test_scaled)[:,1])
roc_auc_score_list.append(roc_auc_score(y_test,y_test_pred))

y_train_pred = random_forest.predict(X_train)
y_test_pred = random_forest.predict(X_test)
fpr_rf, tpr_rf, thresholds = roc_curve(y_test, random_forest.predict_proba(X_test_scaled)[:,1])
roc_auc_score_list.append(roc_auc_score(y_test,y_test_pred))


y_train_pred = extra_tree.predict(X_train)
y_test_pred = extra_tree.predict(X_test)
fpr_etr, tpr_etr, thresholds = roc_curve(y_test, extra_tree.predict_proba(X_test_scaled)[:,1])
roc_auc_score_list.append(roc_auc_score(y_test,y_test_pred))


y_train_pred = adaboost.predict(X_train)
y_test_pred = adaboost.predict(X_test)
fpr_ada, tpr_ada, thresholds = roc_curve(y_test, adaboost.predict_proba(X_test_scaled)[:,1])
roc_auc_score_list.append(roc_auc_score(y_test,y_test_pred))



y_train_pred = gbm.predict(X_train, ntree_limit=gbm.best_ntree_limit)
y_test_pred =gbm.predict(X_test, ntree_limit=gbm.best_ntree_limit)
fpr_gbm, tpr_gbm, thresholds = roc_curve(y_test, gbm.predict_proba(X_test_scaled)[:,1])
roc_auc_score_list.append(roc_auc_score(y_test,y_test_pred))
````

````python
plt.plot(fpr_gbm, tpr_gbm,lw=2,label = "Stacked Voting")
plt.plot(fpr_ada, tpr_ada,lw=2,label = "AdaBoosting")
plt.plot(fpr_etr, tpr_etr,lw=2,label = "Extra Treeing")
plt.plot(fpr_rf, tpr_rf,lw=2,label = "Random Forest")
plt.plot(fpr_logit, tpr_logit,lw=2,label = "Logistic Regression")
plt.plot([0,1],[0,1],c='violet',ls='--')
plt.xlim([-0.05,1.05])
plt.ylim([-0.05,1.05])


plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve for fraud problem');
plt.legend(loc=0)

````
